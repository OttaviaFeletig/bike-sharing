```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
hour_df = pd.read_csv("./hour.csv")
hour_df
```

```python
day_df = pd.read_csv("./day.csv")
day_df[day_df["holiday"] == 1]
```

```python
# CHECKING FOR MISSING VALUES
hour_df.info()
day_df.info()
hour_df.isna().any().any()
hour_df.notna().any().any()
hour_df.isnull().any().any()
day_df.isna().any().any()
day_df.notna().any().any()
day_df.isnull().any().any()
```

```python
# DATA TYPES CONVERSION, CREATING NEW FEATURES AND COLUMN RENAMING
hour_df["dteday"] = pd.to_datetime(hour_df["dteday"])
day_df["dteday"] = pd.to_datetime(day_df["dteday"])

hour_df.loc[hour_df["season"] == 1, "season"] = "winter"
hour_df.loc[hour_df["season"] == 2, "season"] = "spring"
hour_df.loc[hour_df["season"] == 3, "season"] = "summer"
hour_df.loc[hour_df["season"] == 4, "season"] = "fall"
day_df.loc[day_df["season"] == 1, "season"] = "winter"
day_df.loc[day_df["season"] == 2, "season"] = "spring"
day_df.loc[day_df["season"] == 3, "season"] = "summer"
day_df.loc[day_df["season"] == 4, "season"] = "fall"

hour_df.loc[hour_df["yr"] == 0, "yr"] = 2011
hour_df.loc[hour_df["yr"] == 1, "yr"] = 2012
day_df.loc[day_df["yr"] == 0, "yr"] = 2011
day_df.loc[day_df["yr"] == 1, "yr"] = 2012

hour_df.loc[hour_df["holiday"] == 0, "holiday"] = "no"
hour_df.loc[hour_df["holiday"] == 1, "holiday"] = "yes"
day_df.loc[day_df["holiday"] == 0, "holiday"] = "no"
day_df.loc[day_df["holiday"] == 1, "holiday"] = "yes"


# defining if it's weekend or not

day_df["is_weekend"] = day_df["weekday"].apply(lambda value: False if value <= 4 else True)
day_df['is_weekend'] = pd.Categorical(day_df['is_weekend'], categories=[False, True])
hour_df["is_weekend"] = hour_df["weekday"].apply(lambda value: False if value <= 4 else True)
hour_df['is_weekend'] = pd.Categorical(hour_df['is_weekend'], categories=[False, True])

hour_df.loc[hour_df["weekday"] == 0, "weekday"] = "monday"
hour_df.loc[hour_df["weekday"] == 1, "weekday"] = "tuesday"
hour_df.loc[hour_df["weekday"] == 2, "weekday"] = "wednesday"
hour_df.loc[hour_df["weekday"] == 3, "weekday"] = "thursday"
hour_df.loc[hour_df["weekday"] == 4, "weekday"] = "friday"
hour_df.loc[hour_df["weekday"] == 5, "weekday"] = "saturday"
hour_df.loc[hour_df["weekday"] == 6, "weekday"] = "sunday"
day_df.loc[day_df["weekday"] == 0, "weekday"] = "monday"
day_df.loc[day_df["weekday"] == 1, "weekday"] = "tuesday"
day_df.loc[day_df["weekday"] == 2, "weekday"] = "wednesday"
day_df.loc[day_df["weekday"] == 3, "weekday"] = "thursday"
day_df.loc[day_df["weekday"] == 4, "weekday"] = "friday"
day_df.loc[day_df["weekday"] == 5, "weekday"] = "saturday"
day_df.loc[day_df["weekday"] == 6, "weekday"] = "sunday"

hour_df.loc[hour_df["workingday"] == 0, "workingday"] = False
hour_df.loc[hour_df["workingday"] == 1, "workingday"] = True
day_df.loc[day_df["workingday"] == 0, "workingday"] = False
day_df.loc[day_df["workingday"] == 1, "workingday"] = True

# grouping hours into morning, afternoon, evening and night
hour_df["day_time"] = hour_df["hr"].apply(lambda value: "night"
                                            if value <= 5 else "morning"
                                            if value <= 12 else "afternoon"
                                            if value <= 17 else "evening")
hour_df['day_time'] = pd.Categorical(hour_df['day_time'], categories=['night', 'morning', 'afternoon', "evening"])


hour_df.rename(columns={"yr": "year", "mnth": "month", "hr": "hour", "cnt": "total_count"}, inplace=True)
day_df.rename(columns={"yr": "year", "mnth": "month", "cnt": "total_count"}, inplace=True)

def convert_temp(temp):
    return (temp - 32) * (5 / 9)

# changing measure unit for temperature(Â° Celsius), humidity(%) and windspeed(km/h)
hour_df["temp"] = hour_df["temp"].apply(lambda value: convert_temp(value) if value > 1 else convert_temp(value * 100))
hour_df["atemp"] = hour_df["atemp"].apply(lambda value: convert_temp(value) if value > 1 else convert_temp(value * 100))
hour_df["hum"] = hour_df["hum"].apply(lambda value: convert_temp(value) if value > 1 else convert_temp(value * 100))
hour_df["windspeed"] = hour_df["windspeed"].apply(lambda value: convert_temp(value) if value > 1 else convert_temp(value * 100))
day_df["temp"] = day_df["temp"].apply(lambda value: convert_temp(value) if value > 1 else convert_temp(value * 100))
day_df["atemp"] = day_df["atemp"].apply(lambda value: convert_temp(value) if value > 1 else convert_temp(value * 100))
day_df["hum"] = day_df["hum"].apply(lambda value: convert_temp(value) if value > 1 else convert_temp(value * 100))
day_df["windspeed"] = day_df["windspeed"].apply(lambda value: convert_temp(value) if value > 1 else convert_temp(value * 100))

# defining if day is warm or not and if it's rainy

day_df["warm_day"] = day_df["temp"].apply(lambda value: "no" if value < 21 else "yes")
day_df['warm_day'] = pd.Categorical(day_df['warm_day'], categories=['no', 'yes'])
day_df["rainy_day"] = day_df["weathersit"].apply(lambda value: False if value <= 2 else True)
day_df['rainy_day'] = pd.Categorical(day_df['rainy_day'], categories=[False, True])
day_df
```

```python
sns.set(rc={'figure.figsize':(11,8)})

sns.set_context("paper", font_scale=1, rc={"lines.linewidth": 2.5})
#sns.set(style="ticks", context="talk")
#plt.style.use("dark_background")
sns.set_style("dark")

#fig, ax = plt.subplots()
ax = sns.pointplot(data = hour_df[['hour','total_count','season']], x = 'hour', y = 'total_count',
              scale = 0.2, hue = 'season')

plt.setp(ax.get_legend().get_texts(), fontsize='15') # for legend text
plt.setp(ax.get_legend().get_title(), fontsize='16') # for legend title

ax.set(title = 'Season wise hourly distribution of bike rentals',ylabel= 'mean(total_count)');
```

```python
casual_user_workingday_stats = day_df[day_df["workingday"] == True]["casual"].mean()
registered_user_workingday_stats = day_df[day_df["workingday"] == True]["registered"].mean()
casual_user_is_weekend_stats = day_df[day_df["is_weekend"] == True]["casual"].mean()
registered_user_is_weekend_stats = day_df[day_df["is_weekend"] == True]["registered"].mean()
casual_user_is_holiday_stats = day_df[day_df["holiday"] == "yes"]["casual"].mean()
registered_user_is_holiday_stats = day_df[day_df["holiday"] == "yes"]["registered"].mean()
registered_casual_users_stats = pd.DataFrame([
    [casual_user_workingday_stats, "working day", "casual users"], 
    [registered_user_workingday_stats, "working day", "registered users"],
    [casual_user_is_weekend_stats, "weekend", "casual users"], 
    [registered_user_is_weekend_stats, "weekend", "registered users"], 
    [casual_user_is_holiday_stats, "holiday", "casual users"], 
    [registered_user_is_holiday_stats, "holiday", "registered users"]
    ], columns=["amount", "day", "user type"])
ax = sns.barplot(x="day", y="amount", hue="user type", data=registered_casual_users_stats)
ax.set(title = 'Day time amount of bike rentals for casual and registered users',ylabel= 'mean(total_count)');
```
